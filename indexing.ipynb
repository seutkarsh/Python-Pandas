# Indexing


```python
import pandas as pd
```

```python
df = pd.read_csv("world_population.csv")
df
```

## Adding country as index (unique)


```python
#while reading csv
df = df = pd.read_csv("world_population.csv", index_col="Country")
df
```

```python
#after dataframe is read
df = pd.read_csv("world_population.csv")
df.set_index("Country",inplace=True)
df
```

## Resetting Index


```python
df.reset_index(inplace=True)
df
```

## Searching Through Index


```python
df = df = pd.read_csv("world_population.csv", index_col="Country")

# using loc -- location
df.loc["Albania"]
```

```python
# using iloc -- integer location -- integer indexing behind the column value
df.iloc[1]
```

## Creating Multiple Index -- Multi-indexing


```python
df = df = pd.read_csv("world_population.csv")
df.set_index(["Continent","Country"],inplace=True)
df
```

## Searching Multi-Index


```python
# using loc

df.loc["Africa","Angola"]  #  ["Continent","Country"]--indexwise
```

```python
# using Iloc
df.iloc[1]    # this will always go for base indexing..NOT WORKING WITH MULTIPLE INDEXING
```
