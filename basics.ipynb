# Reading Files


```python
 import pandas as pd
```

## CSV Files


```python
# relative path
pd.read_csv("countries of the world.csv")

#absolute path
pd.read_csv(r"D:\Projects\Data Analysis\Pandas\countries of the world.csv")  #used r to make the whole string as raw string hence no unicode error

#if no headers are present
pd.read_csv("countries of the world.csv", header=None, names=["Country", "Region"])
```

## Text Files


```python
# using read_csv ---> without sep defined, comma will be the default sep and hence contents of the file will not separate
df = pd.read_csv("countries of the world.txt", sep='\t')
df
```

```python
# using read_table
df = pd.read_table("countries of the world.txt")  # default sep is tab
df
```

## Json Files


```python
#using read_json
df=pd.read_json("json_sample.json")
df
```

## Excel Files


```python
#using read_excel
df = pd.read_excel("world_population_excel_workbook.xlsx")
df

#this by default reads first sheet in the workbook

#for another sheet
df = pd.read_excel("world_population_excel_workbook.xlsx", sheet_name="Sheet1")
df
```
