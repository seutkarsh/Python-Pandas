# Visualization


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
```

```python
plt.style.use('tableau-colorblind10')
```

```python
df = pd.read_csv("Ice Cream Ratings.csv")
df.set_index("Date",inplace=True)
df
```

```python
# default is "line" plot
df.plot()
```

```python
# for breaking plot column-wise
df.plot(kind="line",subplots=True)
```

```python
# customization
df.plot(kind="line", title="Ice Cream Ratings", xlabel="Daily Rating", ylabel="Scores")
```

```python
# bar plot
df.plot(kind="bar")
```

```python
# stacked bar plot
df.plot(kind="bar", stacked=True)
```

```python
# specific columns
df["Flavor Rating"].plot(kind="bar")
```

```python
# horizontal bar chart
df.plot.barh(stacked=True)
```

```python
#scatter plot
df.plot.scatter(x="Texture Rating", y="Overall Rating")
```

```python
# change marker
df.plot.scatter(x="Texture Rating", y="Overall Rating", s=50, c="Yellow")
```

```python
# histogram
df.plot.hist()
```

```python
#customized hist
df.plot.hist(bins=20)  #bins = x axis tick marks
```

```python
# box plot
df.boxplot()
```

```python
# area plot
df.plot.area()
```

```python
# changing figure size
df.plot.area(figsize=(10,5))
```

```python
#pie chart
df.plot.pie(y="Flavor Rating")
```
